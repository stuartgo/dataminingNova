```python
import pyreadstat
import pandas as pd
df, meta = pyreadstat.read_sas7bdat('a2z_insurance.sas7bdat')
print(df.isna().sum().sum())


def educationToNumber(x):
    if x:
        return int(x[0])
    return -1
#change customerid,firstpolyear,birth year, to integer
df["CustID"]=df["CustID"].astype(int)
df["FirstPolYear"]=df["FirstPolYear"].fillna(-1).astype(int)
df["BirthYear"]=df["BirthYear"].fillna(-1).astype(int)
df["EducDeg"]=df["EducDeg"].fillna(-1).apply(educationToNumber)
df["MonthSal"]=df["MonthSal"].fillna(-1) #should maybe be integer?
df["GeoLivArea"]=df["GeoLivArea"].fillna(-1).astype(int)
df["Children"]=df["Children"].fillna(-1).astype(int)



#Map for EducDeg
#1- Basic
#2- High School
#3- BSc/MSc
#4 PhD
#set CustID to index
df.set_index("CustID")

```
