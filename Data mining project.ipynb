```python
import pyreadstat
import pandas as pd
df, meta = pyreadstat.read_sas7bdat('a2z_insurance.sas7bdat')
print(df.isna().sum().sum())
#just removes all datapoints with Nan, simple soltuion for now
df=df.dropna()


def educationToNumber(x):
    if x:
        return int(x[0])
    return -1

#change customerid,firstpolyear,birth year, to integer
df["CustID"]=df["CustID"].astype(int)
df["FirstPolYear"]=df["FirstPolYear"].astype(int)
df["BirthYear"]=df["BirthYear"].astype(int)
df["EducDeg"]=df["EducDeg"].apply(educationToNumber)
df["MonthSal"]=df["MonthSal"].astype(int) #should maybe be integer?
df["GeoLivArea"]=df["GeoLivArea"].astype(int)
df["Children"]=df["Children"].astype(int)



#Map for EducDeg
#1- Basic
#2- High School
#3- BSc/MSc
#4 PhD
#set CustID to index
df.set_index("CustID")

```
