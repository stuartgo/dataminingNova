```python
import pyreadstat
import pandas as pd
import matplotlib.pyplot as plt
df, meta = pyreadstat.read_sas7bdat('a2z_insurance.sas7bdat')
df.set_index("CustID")
```

```python
#############
#Handling missing values
#############
df=df.loc[~pd.isna(df["FirstPolYear"])] #could try with mode
df=df.loc[~pd.isna(df["BirthYear"])] #could try with mode

df[["MonthSal"]]=df[["MonthSal"]].fillna(df["MonthSal"].mean()) 
df[["PremMotor"]]=df[["PremMotor"]].fillna(df["PremMotor"].mean()) #can try with median as well
df[["PremHealth"]]=df[["PremHealth"]].fillna(df["PremHealth"].median()) #can try with removing as well
df[["PremLife"]]=df[["PremLife"]].fillna(df["PremLife"].mean())
df[["PremWork"]]=df[["PremWork"]].fillna(df["PremWork"].mean()) #can try with median as well
df[["GeoLivArea"]]=df[["GeoLivArea"]].fillna(df["GeoLivArea"].mode())
#TODO
df=df.loc[~pd.isna(df["Children"])]
#need to do knn estimation
#df["Children"] =df["Children"].interpolate()
```

```python
#############
#Cleaning data types
#############


#Map for EducDeg
#1- Basic
#2- High School
#3- BSc/MSc
#4 PhD
#set CustID to index
def educationToNumber(x):
    if x:
        return int(x[0])
    return -1

#change customerid,firstpolyear,birth year, to integer
df["CustID"]=df["CustID"].astype(int)
df["FirstPolYear"]=df["FirstPolYear"].astype(int)
df["BirthYear"]=df["BirthYear"].astype(int)
df["EducDeg"]=df["EducDeg"].apply(educationToNumber)
df["MonthSal"]=df["MonthSal"].astype(int) #should maybe be integer?
df["GeoLivArea"]=df["GeoLivArea"].astype(int)
df["Children"]=df["Children"].astype(int)
```

```python
#############
#Filtering incoherent data
#############
#assume that firstpolyear can be before birthdate, but only up to 18 years, 
#as you are part of your parents policy
df=df[(df["BirthYear"]-df["FirstPolYear"])<18] 

#child at very young age
#age and education
```

```python
#############
#Outlier filtering
#############


df=df[df["FirstPolYear"]<50000]
df=df[df["BirthYear"]>1800]
df=df[df["MonthSal"]< 30000]
df=df[df["CustMonVal"]> -100000]
df=df[df["ClaimsRate"]< 50]
df=df[df["PremMotor"] < 2000] #needs to be tested, can be removed
df=df[df["PremMotor"] < 3000] #needs to be tested, can be removed
df=df[df["PremWork"] < 750] #needs to be tested, can be removed
df.shape
```
